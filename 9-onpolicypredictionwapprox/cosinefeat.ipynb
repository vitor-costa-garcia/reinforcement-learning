```python
import numpy as np
import matplotlib.pyplot as plt

s1 = np.arange(0, 1, 0.01)
s2 = np.arange(0, 1, 0.01)

c = np.array([[0, 1], [1, 0], [1, 1], [0, 5], [2, 5], [5, 2]])

S1, S2 = np.meshgrid(s1, s2, indexing='ij')

fig, ax = plt.subplots(2, 3, figsize=(12, 8))

for idx, vec in enumerate(c):
    M = np.cos(np.pi * (S1 * vec[0] + S2 * vec[1]))

    ax[idx // 3, idx % 3].imshow(M, cmap='gray', origin='lower', extent=[0,1,0,1])
    ax[idx // 3, idx % 3].set_title(f"c = {vec}")
    ax[idx // 3, idx % 3].axis('off')

plt.tight_layout()
plt.show()
```
